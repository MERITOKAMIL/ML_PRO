Simplifier le tout avec pipeline

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer

print(f"Python version : {sklearn.__version__}")
```

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, random_state=0)
```

```python
pipeline = Pipeline(steps=[("encoder", OrdinalEncoder()),("normaliser", MinMaxScaler())])
```

```python
pipeline
```

Column transfomer

```python
categorical_col=["cut", "color", "clarity"]
```

```python
df[categorical_col].isna().sum()
```

```python
column_transformer = ColumnTransformer(transformers=[("encoder", OrdinalEncoder()), categorical_col], remainder="passthrough")
column_transformer
```

```python
pipeline = make_pipeline(column_transformer, MinMaxScaler())
```

```python
pipeline.fit_transform(train_set)
```
