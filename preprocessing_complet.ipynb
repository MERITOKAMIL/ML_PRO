```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, MinMaxScaler,PolynomialFeatures,StandardScaler
from sklearn.feature_selection import SelectKBest, chi2,f_regression,f_classif
```

**DATA**

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.isna().sum(axis=0)
```

```python
df.info()
```

```python
df_cat=df.select_dtypes(include="category")
col_cat=list(df_cat.columns)
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoder_result = encoder.fit_transform(df[col_cat])
encoder_result
```

```python
encoded_columns = encoder.get_feature_names_out(col_cat)
encoded_df = pd.DataFrame(encoder_result, columns=encoded_columns, index=df.index)
encoded_df[encoded_columns]
```

```python
X=encoded_df[encoded_columns]
y=df[["smoker"]]
```

```python
selector=SelectKBest(f_classif,k=5)
selector.fit(X,y)
selector.get_feature_names_out()
```

```python
selected_df=X[selector.get_feature_names_out()]
selected_df
```

```python
df_num=df.select_dtypes(exclude="category")
df_num
```

```python
df_num.columns
```

```python
X_num=df_num[[ "size"]]
y_num=df_num[["total_bill"]]
```

```python
Scaler=MinMaxScaler()
Scaler.fit(X_num)
X_num_scaled=Scaler.transform(X_num)
X_num_scaled
```

```python
stand_scaler=StandardScaler()
stand_scaler.fit(y_num)
y_num_stand=stand_scaler.transform(y_num)
y_num_stand
```

```python
df_num_scaled=pd.DataFrame(X_num_scaled,columns=X_num.columns)
df_num_scaled
df_num_scaled["total_bill"]=y_num_stand
df_num_scaled
```

```python
#concaténation
df_final=pd.concat([selected_df,df_num_scaled],axis=1)
df_final
```

```python
df_final.isna().sum(axis=0)
```

```python
polynomial_final=PolynomialFeatures(degree=2, include_bias= False)
polynomial_final.fit(df_final)
polynomial_final.get_feature_names_out()
```

```python
df_poly=pd.DataFrame(polynomial_final.transform(df_final),columns=polynomial_final.get_feature_names_out())
df_poly
```

```python
ploy_scaler=MinMaxScaler()
ploy_scaler.fit(df_poly)
df_poly_scaled=ploy_scaler.transform(df_poly)
df_poly_scaled = pd.DataFrame(df_poly_scaled, columns=df_poly.columns)
df_poly_scaled
```
